```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../../../data/struc/demo.csv')
df.head()
```

```python
''' plot age sex education '''
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.histplot(df['age'], ax=ax[0])
ax[0].set_title('Age')
sns.countplot(df['sex'], ax=ax[1])
ax[1].set_title('sex')
sns.histplot(df['edu'], ax=ax[2])
ax[2].set_title('education')
```

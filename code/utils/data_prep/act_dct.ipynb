# visualize the count for each number in list 'numbers' for dict extracted from json `res/act_dict_extend.json`

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dic = json.load(open('res/act_dict_extend.json'))
dic
```

```python
#iterate over the dic counting how many times each numer 1-9 appears in the list 'numbers' for each item
count = {k: {str(i): 0 for i in range(1, 10)} for k in dic.keys()}
for k, v in dic.items():
    for n in v['numbers']:
        count[k][str(n)] += 1

#convert the count dict to a pandas dataframe
df = pd.DataFrame(count).T
```

```python
df
```

```python
#sum these up and visualize in a barplot
df.sum().plot(kind='bar')
plt.show()
```
